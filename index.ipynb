```python
import re

def get_super_built_up_area(text):
    match = re.search(r'Super Built up area (\d+\.?\d*)', text)
    if match:
        return float(match.group(1))
    return None

# Function to extract the Built-up or Carpet area
def get_area(text, area_type):
    match = re.search(rf'{area_type}\s*:\s*(\d+\.?\d*)', text)
    if match:
        return float(match.group(1))
    return None

# Function to check if the area is in sq.m. and convert it to sqft
import re

def convert_to_sqft(text, area_value):
    if area_value is None:
        return None
    match = re.search(rf'{area_value} \((\d+\.?\d*) sq\.m\.\)', text)
    if match:
        sq_m_value = float(match.group(1))
        return round(sq_m_value * 10.7639, 1)  # Rounding to one decimal place
    return round(area_value, 1) if area_value is not None else None
```

```python
def extract_property_type(property_name):
    if 'House' in property_name:
        return 'House'
    elif 'Flat' in property_name or 'BHK' in property_name:
        return 'Flat'
    return 'Unknown'

# Apply the function to create a new column 'property_type'
```

```python
df['property_type'] = df['property_name'].apply(extract_property_type)
```

```python
df['super_built_up_area'] = df['areaWithType'].apply(lambda x: get_super_built_up_area(str(x)))
df['super_built_up_area'] = df.apply(lambda x: convert_to_sqft(str(x['areaWithType']), x['super_built_up_area']), axis=1)
```

```python
# extract the built-up area
df['built_up_area'] = df['areaWithType'].apply(lambda x: get_area(str(x), 'Built Up area'))
df['built_up_area'] = df.apply(lambda x: convert_to_sqft(str(x['areaWithType']), x['built_up_area']), axis=1)
```

```python
#  extract the carpet area
df['carpet_area'] = df['areaWithType'].apply(lambda x: get_area(str(x), 'Carpet area'))
df['carpet_area'] = df.apply(lambda x: convert_to_sqft(str(x['areaWithType']), x['carpet_area']), axis=1)
```
